# BUANG OUTLIER

```python
#Q1, Q3, dan IQR
kolom = ["Balance", "EstimatedSalary"]

Q1 = parisya[kolom].quantile(0.25) #menghitung kuartil pertama (Q1) untuk setiap kolom yang ada di dalam list kolom
Q3 = parisya[kolom].quantile(0.75) #menghitung kuartil ketiga (Q3) untuk setiap kolom yang ada di dalam list kolom
IQR = Q3-Q1 #enghitung Rentang Interkuartil (IQR) untuk setiap kolom yang ada di dalam list kolom
parisya = parisya[~((parisya[kolom]<(Q1 - 1.5 * IQR)) |
             (parisya[kolom]>(Q3 + 1.5 * IQR))).any(axis = 1)]

plt.boxplot(parisya[kolom]) #membuat boxplot dari data yang sudah bersih dari outlier
plt.xticks([1,2], kolom) #menetapkan label sumbu x untuk kolom yang sesuai
plt.title("Outlier Setelah Dibuang") #menetapkan judul untuk boxplot
plt.show() #menampilkan data
```

# MEMBUAT NILAI X

```python
x_array = np.array(parisya) 
```

# VISUALISASI PERSEBARAN DATA

```python
plt.scatter(parisya.Balance, parisya.EstimatedSalary)
plt.show()
```

# STANDARISASI DATA

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_array)
plt.scatter(x_scaled[:,0], x_scaled[:,1], s=10)
plt.show()
```

# CEK SUM OF SQUARE ERROR DARI TIAP PEMBAGIAN JUMLAH CLUSTER

```python
from sklearn.cluster import KMeans
sse = []
index = range(1,10)
for i in index :
    kmeans = KMeans(n_clusters = i, random_state = 0, n_init='auto')
    kmeans.fit(x_scaled)
    sse_ = kmeans.inertia_
    sse.append(sse_)
    print (i,sse_)
```

```python
plt.plot(index, sse)
plt.show() 
```

# MEMBUAT MODEL

```python
kmeans = KMeans (n_clusters = 3, random_state = 0, n_init='auto')
kmeans.fit(x_scaled)
```

# MELIHAT CLUSTER PUSAT

```python
kmeans.cluster_centers_
```

# VISUALISASI PERSEBARAN DATA SETELAH CLUSTERING

```python
output = plt.scatter(x_scaled[:,0],x_scaled[:,1], s=10, c = kmeans.labels_) #DATANYA

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c="orange", s=10) #centroid
plt.title("KMEANS PREDIKSI CHURN NASABAH BANK")
plt.xlabel("Balance")
plt.ylabel("EstimatedSalary")
plt.colorbar(output)
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score
labels = kmeans.labels_
davies_bouldin_score(x_scaled, labels)
```
